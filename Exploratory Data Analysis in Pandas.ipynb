```python
# Check the data types of all columns
print(df.dtypes)

# Attempt to convert each column to numeric
for column in df.columns:
    try:
        df[column] = pd.to_numeric(df[column])
    except ValueError as e:
        print(f"Error converting column {column}: {e}")
```
